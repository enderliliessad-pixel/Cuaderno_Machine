```python
# Crea una función en la cual reciba 3 parametros y verifique cual de ellos es el numero mayor
def numero_Mayor(num1,num2,num3):
    if(num1 > num2 and num1 > num3):
        print("El numero mayor es", num1)
    else:
        if(num2>num1 and num2 > num3):
            print("El numero mayor es", num2)
        else:
            print("El numero mayor es", num3)
```

```python
num1=int(input("Coloque un numero"))
num2=int(input("Coloque un segundo numero"))
num3=int(input("Coloque un tercer numero"))

numero_Mayor(num1,num2,num3)
```

## **Que es un vector**
Es un espacio de la memoria que se encarga de guardar varios datos

```python
# Definimos el vector vacio
lista=[]
tamanio=int(input("Coloque el tamaño del vector"))
for i in range(tamanio):
    n=int(input("Ingrese un valor entero"))
    lista.append(n)
#Imprimir lo que tiene la lista
print (lista)
```

```python
# Crear una función donde se ingrese o se genere un vector con número aleatorio del o al 10
import random as rd
def crearVector(n):
    vector=[]
    for i in range(n):
        numero=rd.randint(0,10)
        vector.append(numero)
    return vector
```

```python
n= int(input("Coloque un numero"))
crearVector(n)
```

```python
# Crear un función que reciba el tamaño del vector e ingrese el vector por teclado
def t_vector(n):
    vector=[]
    for i in range(n):
        n=int(input("Ingrese los valores del vector"))
        vector.append(n)
    return(vector)
```

```python
n=int(input("Ingrese el tamaño que tendra el vector"))
t_vector(n)
```

```python
def leerVector(n):
    vector=[]
    for i in range(n):
        elemento=int(input("Ingrese el valor{ i + 1 } del vector"))
        vector.append(elemento)
    return vector
    
def mayor(vector):
    maximo=vector[0]
    for i in vector:
        if i > maximo:
            maximo = i
    return maximo
```

```python
n = int(input("Ingrese el tamanio del vector"))
vector = leerVector(n)
maximo = mayor(vector)
print("Vector : ", vector)
print (" El numero mayor es ", maximo)
```
