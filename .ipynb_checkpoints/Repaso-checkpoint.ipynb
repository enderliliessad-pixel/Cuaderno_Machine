```python
"""
3. Escribe un programa que, dado dos strings S1 y S2 y dos números enteros N1 y N2, determine si el substring que en S1 se extiende desde la posición N1 a la N2 (ambos inclusive) está contenido en S2.
"""
s1="Hornitorinco"
s2="Gato"
n1=4
n2=6

fragmento=s1[n1:n2+1]
print("El substring es:", fragmento)
print("EL substring aparece en la otra", fragmento in s2)
```

```python
Lista_Repetida=[1,2,3,4,5,6,7,9,1,2,34,5,7,8,2,3,5,67,9,4,12,3,5,67,87,34,12]

Lista_Sinrepetidos=(set(Lista_Repetida))

print(Lista_Sinrepetidos)


Lista_God=[]

for i in Lista_Repetida:
    if i not in Lista_God:
        Lista_God.append(i)
print(Lista_God)
```

```python
Lista=["Carro", "Mochila","Cuero","Sombrero"]
s="Sombrero"
t="Avion"

for i in range (len(Lista)):
    if Lista[i] == s:
        Lista[i] = t
print(Lista)
```
