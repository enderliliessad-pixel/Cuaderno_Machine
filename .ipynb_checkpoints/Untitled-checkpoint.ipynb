```python
import numpy as np
import math

# Sistema A
A = np.array([[-9, 4],
              [16, -7]])
b = np.array([20, 80])

x, y = np.linalg.solve(A, b)
print(x, y)

# Verificación (debe dar True)
print(math.isclose(-9*x + 4*y, 20))
print(math.isclose(16*x - 7*y, 80))
```

```python
import math
import numpy.linalg as linalg

a = np.array([ [-9,4], [16,-7] ])
b = np.array([20,80])

print(a)
print(b)
x,y = linalg.solve(a,b)
print(f'x={x} y={y}')

print(math.isclose(-9*x + 4*y, 20))
print(math.isclose(16*x - 7*y, 80))

A2 = np.array([[1, -2, 3], [2, 1, 1], [-3, 2, -2]])
b2 = np.array([7, 4, -10])

x,y,z = np.linalg.solve(A2, b2)
print(f'x={x} y={y} z={z}')

print(math.isclose(+1*x-2*y+3*z,7))
print(math.isclose(2*x+1*y+z,4))
print(math.isclose(-3*x+2*y-2*z,-10))
```

```python
import numpy as np

def verificar_A(x, y):
    return (-9*x + 4*y, 16*x - 7*y)

A = np.array([[-9, 4],
              [16, -7]])
b = np.array([20, 80])

x, y = np.linalg.solve(A, b)

print(f"Solución: x={x:.4f}, y={y:.4f}")
izq1, izq2 = verificar_A(x, y)
print("Comprobación:", izq1, "≈ 20 y", izq2, "≈ 80")
```

```python
import pandas as pd

s = pd.Series([1, 2.5, "hola", True, 99, "AI", 0.1, False, "Python", 7],
              index=list("abcdefghij"))

# Por posición: elementos 2,3,4,5 (4 elementos)
parte_iloc = s.iloc[2:6]

# Por etiqueta: desde 'c' hasta 'f' (incluye 'f')
parte_loc = s.loc["c":"f"]

print(parte_iloc)
print(parte_loc)

print("¿Son iguales?", parte_iloc.equals(parte_loc))
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.randint(0, 21, size=(10, 10)))

# Reemplazar < 5 por NaN
df = df.mask(df < 5, np.nan)

# Contar NaN por fila
nans_por_fila = df.isna().sum(axis=1)

print(df)
print("\nNaN por fila:")
print(nans_por_fila)
```

```python
import numpy as np

A = np.array([[4,2],
              [1,3]])

valores, vectores = np.linalg.eig(A)

print("Valores propios:", valores)
print("Vectores propios:\n", vectores)
```

```python
v = vectores[:,0]      # primer vector propio
lam = valores[0]      # primer valor propio

print(A @ v)
print(lam * v)
```

```python
import numpy as np

A = np.array([[3,1],
              [1,3]])

vals, vecs = np.linalg.eig(A)

for i in range(len(vals)):
    v = vecs[:,i]
    lam = vals[i]

    izquierda = A @ v
    derecha = lam * v

    print(f"\nEigenvalor {i+1}:", lam)
    print("Av =", izquierda)
    print("λv =", derecha)
```
