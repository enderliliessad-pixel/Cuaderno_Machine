### 3.2 ¿Qué es un intérprete?
* Es un software que puede leer y ejecutar diferentes instrucciones de un lenguaje de programacion.
* Una capa de software entre el código y el hardware: El interprete funciona como un intermediario pues recive el codigo, lo traduce a acciones que puede entender el sistema y cuando se van a ejecutar dichas acciones.
* Para poder ejecutar programas de Python se necesita tener instalado el interprete en tu computadora. <br>
Para las especificaciones <a href="https://docs.python.org/3/reference/index.html">especificaciones de Python</a> existen varias implementaciones: 
  * CPython: Es la mas comun y es una implementación de C
  * Jython: Es una implementación de Java
  * IronPython: Es una implementación de .NET
