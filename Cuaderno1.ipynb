### 3.2 ¿Qué es un intérprete?
* Es un software que puede leer y ejecutar diferentes instrucciones de un lenguaje de programación.
* Una capa de software entre el código y el hardware: El interprete funciona como un intermediario pues recibe el codigo, lo traduce a acciones que puede entender el sistema y luego se van a ejecutar dichas acciones.
* Para poder ejecutar programas de Python se necesita tener instalado el intérprete en tu computadora. <br>
Para las especificaciones <a href="https://docs.python.org/3/reference/index.html">especificaciones de Python</a> existen varias implementaciones: 
  * CPython: Es la mas comun y es una implementación de C
  * Jython: Es una implementación de Java
  * IronPython: Es una implementación de .NET

### 3.5 Perspectiva de Python
1. Compilación de código fuente a bytecode: En Python el código escrito no se ejecuta directamente sino que
  * El bytecode se ejecuta más rápido porque no se debe analizar cada vez
  * El bytecode se guarda en archivos con la extensión .pyc que se almacenan en caché.
  * Si el código fuente no se modificó se usara el bytecode que le permite saltarse el paso de compilación.
2. Máquina virtual de Python (PVM)
   * Ejecuta las instrucciones en bytecode <br>

### 3.8 REPL
El Repl es un sistema interactivo que permite la comunicación con el ordenador al usar un lenguaje de programación como Python, y el Relp se basa en cuatro pasos fundamentales:
* Read: El sistema recibe y lee las instrucciones que el usuario escribe
* Evaluate: La instrucción es analizada y ejecutada por el intérprete
* Print: El resultado se muestra en pantalla
* Loop: El proceso se repite sin la necesidad de reiniciar el sistema <br>
