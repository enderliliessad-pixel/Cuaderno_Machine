```python
arr = np.empty((3,3))

for i in range(3):
    for j in range(3):
        arr[i,j] = i*j

print(arr)
```

```python
## Find indices of non-zero elements from [1,2,0,0,4,0]
lista=[1,2,0,0,4,0]

listan=[]

for x in range(len(lista)):
    if lista[x] != 0:
        listan.append(x)

print (listan)
```
